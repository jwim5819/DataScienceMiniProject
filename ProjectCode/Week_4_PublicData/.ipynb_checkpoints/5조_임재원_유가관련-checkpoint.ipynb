# <span style='color:Blue'>[주제]  원유 가격 상승이 국내 기업들에 미친 영향을 알아보고 분석한다.</span>

### [서론]
· 전 세계가 코로나 19의 공포에서 회복되며 원유 수요가 급증한 한편, 러시아의 우크라이나 침공으로 미국의 러시아산 원유 수입 금지령과 더불어 원유 공급이 부족해질 것이라는 공포감의 확산은 원유 가격의 폭등으로 이어지게 되었다.  
· 따라서 이러한 유가폭등이 국내 기업들에게 미친 영향을 알아보고 국내산업이 나아가야할 방향성을 제시한다.

### [목차]
(1) 국제 유가 변동 분석  
(2) 국내 유가 변동 분석

(7) 데이터 출처

```python
# 모듈 삽입
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import platform
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False

# Data 관련 변수 선언
DIR_PATH = r'./data/'
```

## <span style='color:Blue'>(1) 국제 유가 변동 분석</span>

```python
# 국제 석유제품 가격 data 읽어오기 및 dataframe확인
oilDF = pd.read_csv(DIR_PATH+'2020.01~2022.07 국제 석유제품가격(월간).csv', encoding='euc-kr')
oilDF.head()
```

```python
# 데이터 전처리    * 재사용 할 일이 없기때문에 inplace=True로 설정
# (1) '기간' column 인덱스로 설정
oilDF.set_index('기간', inplace=True)

# (2) 필요없는 column인 '나프타' 삭제
oilDF.drop('나프타', axis=1, inplace=True)

# (3) column명 재설정
oilDF.rename(columns={'휘발유(92RON)':'휘발유','경유(0.05%)':'경유'}, inplace=True)

# (4) dataframe 확인
oilDF.head()
```

```python
# 국제 유가 변동현황 그래프로 시각화
# (1) '기간' index를 x축으로 사용하기 위해 list화
x = oilDF.index.to_list()

# (2) 각각의 column들을 y축으로 사용하기 위해 list화
y_1 = oilDF['휘발유'].to_list()
y_2 = oilDF['등유'].to_list()
y_3 = oilDF['경유'].to_list()

# (3) 그래프로 시각화
if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='AppleGothic')
plt.figure(figsize=(20,7))
plt.title('국제 석유제품 가격 변화 추이', size=30)
colors = ['#3366CC','#333333','#FF6666']
plt.plot(x,y_1, color=colors[0], label='휘발유', marker='s', markersize=8, linestyle='--', linewidth=1.0)
plt.plot(x,y_2, color=colors[1], label='등유', marker='o', markersize=8, linestyle='--', linewidth=1.0)
plt.plot(x,y_3, color=colors[2], label='경유', marker='^', markersize=8, linestyle='--', linewidth=1.0)
plt.xticks(ticks=x, labels=x, rotation=45, size=15)
plt.legend(fontsize=25)
plt.annotate('러·우 전쟁', ha = 'center', va = 'bottom', xytext = (x[21],y_1[30]), xy = (x[25], y_1[25]),
arrowprops = { 'facecolor' : 'r', 'edgecolor':'b', 'shrink' : 1}, size=23)
plt.style.use('ggplot')
plt.show()
```

### 분석 : 20년 04월 이후 꾸준한 상승세를 보이며, 22년 6월에 모든 유류 최고치를 찍음

## <span style='color:Blue'>(4) 원유 가격 상승에 대한 악영향을 받은 국내 기업 분석</span>
- 유가 민감업종을 중심으로 분석해 보았음
- 부문 : 석유화학, 반도체, 섬유, 철강, 자동차

```python
# exel에서 원하는 sheet읽어서 원하는 데이터만 뽑아오는 함수 생성
# 매출, 영업이익 추출 후 영업이익률 계산
# 영업이익률 = 영업이익 / 총매출
def read_exel_data(file_name):
    dataDF = pd.read_excel(DIR_PATH+file_name, sheet_name='손익계산서', header=6)
    sales_price = dataDF['3개월'].iloc[0]
    sales_profit = dataDF['3개월'].iloc[4]
    profit_margin = sales_profit / sales_price
    return profit_margin

def read_exel_data2(file_name):
    dataDF = pd.read_excel(DIR_PATH+file_name, sheet_name='포괄손익계산서', header=6)
    sales_price = dataDF['3개월'].iloc[0]
    sales_profit = dataDF['3개월'].iloc[4]
    profit_margin = sales_profit / sales_price
    return profit_margin
```

```python
# 회사별 그래프 시각화 함수 생성
def company_graph(company, company_data, field, color):
    if platform.system() == 'Windows':
        plt.rc('font', family='Malgun Gothic')
    else:
        plt.rc('font', family='AppleGothic')
    title = company + ' 영업이익률' + ' [' + field + ']'
    plt.figure(figsize=(20,7))
    plt.title(title, size=30)
    colors = ['#3366CC','#333333','#FF6666']
    plt.plot(date_list,company_data, color=colors[color], label='영업이익률', marker='o', markersize=10, linestyle='--', linewidth=2.0)
    plt.xticks(ticks=date_list, labels=date_list, size=15)
    plt.legend(fontsize=25)
    plt.style.use('ggplot')
    plt.show()
```

## 롯데케미컬[석유화학]

```python
# (1) 롯데케미칼 데이터 list화
date_list = ['20.하반기','20.상반기','21.하반기','21.상반기','22.하반기']
lotte_profit = []
for i in range(1,6):
    file_name = '롯데케미칼' + str(i) + '.xls'
    lotte_profit.append(read_exel_data2(file_name))

# (2) 롯데케미칼 영업이익률 그래프로 시각화
company_graph('롯데케미컬', lotte_profit, '석유화학', 0)
```

## 유가 민감업종 영업이익률 종합

```python
# 종합 비교를 위한 데이터 정규화 작업(standard scale) 모듈 로딩
from sklearn.preprocessing import StandardScaler

# (1) Dataframe으로 종합
all_data = pd.DataFrame([lotte_profit, sk_profit, hyundai_profit, hyundai_car_profit, ko_profit]).T
all_data.columns = ['석유화학','반도체','철강','자동차','섬유산업']

# (2) 정규화 함수 생성
def nomalization(DF_name, col_name):
    sdscaler = StandardScaler()
    sdscaler.fit(DF_name[[col_name]])
    result = sdscaler.transform(DF_name[[col_name]])
    return result

# (3) 그래프로 시각화
if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='AppleGothic')
plt.figure(figsize=(20,7))
plt.title('유가 민감업종 영업이익률 종합', size=30)
colors = ['#3366CC','#FF6666','#333333','#336600', '#6600CC']
plt.plot(date_list,nomalization(all_data, '석유화학'), color=colors[3], label='석유화학', marker='s', markersize=8, linestyle='--', linewidth=2.0)
plt.plot(date_list,nomalization(all_data, '반도체'), color=colors[2], label='반도체', marker='^', markersize=8, linestyle='--', linewidth=2.0)
plt.plot(date_list,nomalization(all_data, '철강'), color=colors[1], label='철강', marker='o', markersize=8, linestyle='--', linewidth=2.0)
plt.plot(date_list,nomalization(all_data, '자동차'), color=colors[0], label='자동차', marker='*', markersize=10, linestyle='--', linewidth=2.0)
plt.plot(date_list,nomalization(all_data, '섬유산업'), color=colors[4], label='섬유산업', marker='s', markersize=10, linestyle='--', linewidth=2.0)

plt.xticks(ticks=date_list, labels=date_list, size=15)
plt.legend(fontsize=25, loc=2)
plt.style.use('ggplot')
plt.show()
```

### 분석 : 석유 의존도가 높은 산업에서 유가상승으로 인한 영업 이익률 하락세를 보이는 분야가 있었음
- 석유화학 : 기초원료인 원유가격 폭등으로 채산성 악화가 불가피함
- 반도체 : 유가상승으로 인한 국제수급동향이 소극적으로 변화하면서 매출에 영향을 준것으로 보임
- 섬유산업 : 석유화학과 마찬가지로 섬유의 원료 대부분이 원유에서 추출되므로 채산성 악화가 불가피함
- 철강, 자동차 : 예상보다는 유가상승으로 인한 영업이익률의 하락은 보이지 않았으나, 22년 상반기의 재무제표 데이터가 없어서 섣불리 판단할 수 없음

## <span style='color:Blue'>(6) 최종 정리</span>
· 2020년 조사결과에 따르면 한국의 원유 의존도가 OECD 37개 회원국들 중 가장 높다고 한다. 이는 석유화학, 철강, 자동차 등 유가 민감산업이 한국 경제에서 차지하는 비중이 크고 원유를 수입에만 의존하기 때문이다. 이는 유가가 상승할때 다른국가에 비해 인플레이션의 압력이 한국에 더 크게 작용하는 결과를 가져온다.  
&nbsp;최근 유가 폭등으로 원유 민감도가 높은 우리나라 주요 산업분야에서 영업이익이 떨어졌을 것이라는 가정하에 데이터를 분석했고, 실제로 대부분의 주요 산업분야에서 예상과 같은 결과를 볼 수 있었다. 하지만 해양플랜트의 비중이 높은 조선업과 같이 유가상승에 대하여 긍정적인 영향을 받는 분야도 있음을 알게 되었다.  
&nbsp;허나, 그러한 분야도 있다는 것이지 여전히 대부분의 우리나라 기반 산업은 유가상승에 따라 타격을 받을 수 밖에 없다. 최근의 전쟁이 아니더라도 과거의 1, 2차 오일쇼크 등 원유값의 폭등은 여러차례 있어 왔고, 미래에도 이러한 유가폭등 사태가 없을 것이라 단언할 수 없다.  
&nbsp;높은 원유 의존도를 가지는 산업구조는 하루아침에 바뀔 수 있는 것이 아니지만, 미래의 오일쇼크를 대비하기 위해 원유의 안정적 공급망 확보 및 자체적인 원유 비축에 힘을 써야 할것이고, 장기적으로는 화석연료의 의존도를 낮추는 쪽으로의 산업구조 개선이 필요하다고 생각한다.
