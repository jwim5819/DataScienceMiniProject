# 구글 영화 검색량이 늘어났던 날 일어난 일

```python
#뉴스의 제목, 링크, 날짜를 크롤링하는 함수

def find_movie_news(url):
#     url = "https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EB%B6%84%EB%8D%B0%EC%8A%A4%EB%A6%AC%EA%B0%80&sort=0&photo=0&field=0&pd=3&ds=2011.01.08&de=2012.02.08&cluster_rank=22&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:from20110108to20120208,a:all&start=11"
#    print(url)
    response = requests.get(url)
    response.raise_for_status()
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')

    # print(soup)

    pos = "#main_pack > section.sc_new.sp_nnews._prs_nws > div > div.group_news > ul"
    ul = soup.select_one(pos)

    # print(ul) 
    lis = ul.find_all("li", class_="bx")
    

#    print(len(lis))
    for i, tt in enumerate(lis):
        print (i+1)
    # tt = lis[1]

        temp = tt.find("div", class_="news_info").find("div", class_="info_group" )  # date
        movie_news_date = temp.find("span", class_="info").text
        
        news_title = tt.find("div", class_="news_area").find("a", class_="news_tit").text
        title

        link = tt.find("div", class_="news_area").find("a", class_="news_tit")["href"]
        link


        print(movie_news_date+"|"+news_title+"|"+link)
        data = movie_news_date + news_title + link
        
    return data   
```

```python
import requests
from bs4 import BeautifulSoup

search_word = input("검색어 :")
start_page = int(input("첫페이지 :"))
end_page = int(input("끝페이지 :"))
start_year=input("시작 년도_8자리 :")
year_end=input("끝 년도_8자리 :")

print("[검색어] '"+search_word+"' [검색페이지]" + str(start_page)+"부터"+str(end_page)+"까지"+ start_year + year_end+'\n')

base_url = "https://search.naver.com/search.naver?where=news&query="

for i in range(start_page, end_page+1):
    url= base_url+str(search_word)+"&sm=tab_opt&sort=0&photo=0&field=0&pd=3&ds="+start_year+"&de="+year_end+"&docid=&related=0&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so%3Ar%2Cp%3Afrom"+start_year+"to"+year_end+"a:all&start="+str(10*i-9)
    find_movie_news(url)
```
